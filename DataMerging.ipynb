```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Save file path to variable
data20172 = "2017.2.csv"
data20173 = "2017.3.csv"
data20174= "2017.4.csv"
data20181= "2018.1.csv"
data20182= "2018.2.csv"
data20183 = "2018.3.csv"
data20184 = "2018.4.csv"
data20191= "2019.1.csv"
```

```python
# Read with Pandas
pokemon_pd = pd.read_csv(data20172)
pokemon_pd.head()
```

```python
#pokemon_pd.info
pokemon_pd =pokemon_pd.append(pd.read_csv(data20173))
```

```python
pokemon_pd=pokemon_pd.append(pd.read_csv(data20174))
pokemon_pd=pokemon_pd.append(pd.read_csv(data20181))
pokemon_pd=pokemon_pd.append(pd.read_csv(data20182))
pokemon_pd=pokemon_pd.append(pd.read_csv(data20183))
pokemon_pd=pokemon_pd.append(pd.read_csv(data20184))
pokemon_pd=pokemon_pd.append(pd.read_csv(data20191))
```

```python
# Create new table
pokemon_type = pokemon_pd[["Type 1", "HP", "Attack",
                           "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_type.head()
```

```python
pokemon_pd.info
```

```python
# Create the GroupBy object based on the "Type 1" column
pokemon_group = pokemon_type.groupby(["Type 1"])

# Calculate averages for combat stats using the .mean() method
pokemon_comparison = pokemon_group.mean()
pokemon_comparison
```

```python
# Total number of points
pokemon_comparison["Total"] = pokemon_comparison["HP"] + pokemon_comparison["Attack"] + pokemon_comparison["Defense"] + \
    pokemon_comparison["Sp. Atk"] + \
    pokemon_comparison["Sp. Def"] + pokemon_comparison["Speed"]

pokemon_comparison["Total"]
```

```python
# Sort by strongest Pokemon, and reset index
strongest_pokemon = pokemon_comparison.sort_values(["Total"], ascending=False)
strongest_pokemon.reset_index(inplace=True)
strongest_pokemon
```

```python
# Save output to Excel
pokemon_pd.to_csv("24Mo_Compiled_Citibike_Data.csv", index=False)
```
